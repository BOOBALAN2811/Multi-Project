```python
#importing the beautiful soup
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
titles=soup.find_all('ul',class_="vector-menu-content-list")
titles
```

```python
Header=[]
for i in titles:
    Header.append(i.text.replace('\n',' '))
Header
```
