```python
#importing the beautiful soup
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
title2=soup.find_all('tr',class_="table-body")
ICC_Men=[]
for i in title2:
    ICC_Men.append(i.text.replace('\n',' '))
print(ICC_Men)
```
