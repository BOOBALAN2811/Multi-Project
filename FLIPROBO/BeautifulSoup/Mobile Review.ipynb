```python
#importing the beautiful soup
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.amazon.in/s?k=mobile+under+20000&crid=233TS7GZW6CUO&sprefix=mobile+under+2%2Caps%2C421&ref=nb_sb_ss_ts-doa-p_1_14")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
title1=soup.find_all('h2',class_="a-size-mini a-spacing-none a-color-base s-line-clamp-2")
title2=soup.find_all('span',class_="a-price-whole")
title3=soup.find_all('i',class_="a-icon a-icon-star-small a-star-small-4 aok-align-bottom")
```

```python
Mobile_Name=[]
for i in title1:
    Mobile_Name.append(i.text.replace('\n',' '))
print(Mobile_Name)

Price=[]
for i in title2:
    Price.append(i.text.replace('\n',' '))
print(Price)

Rating=[]
for i in title3:
    Rating.append(i.text.replace('\n',' '))
print(Rating)
```
